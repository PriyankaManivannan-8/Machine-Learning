### 3.	Dimensionality Reduction (15 marks)

a.	How will you decide when to apply PCA based on the correlation? (2 marks) 

b.	Apply PCA on the above dataset and determine the number of PCA components to be used so that 90% of the variance in data is explained by the same. (7 marks)

c.	Build a data frame with the principal components and check if multi-collinearity still exists. ( 2  marks).

d.	Visualize the spread of data across PCA components. (2 marks)

e.	Check for outliers in the PCA data and treat the same. (2 marks)


```python
#B)
# Step - 1) Scaling. Since scaling is already perfromed, skipping it.
# Step - 2) Calculating cov matrix:
cov_matrix=np.cov(df_sc.T)
print('Covariance Matrix \n',cov_matrix)
```

```python
# Step - 3) Fing Eigen value and Eigen Vectors:
eig_vals,eig_vecs=np.linalg.eig(cov_matrix)
print('Eigen Vectors \n',eig_vecs)
print('\nEigen Values \n',eig_vals)
```

```python
# Step - 4) Explained Variance Ratio:
tot=sum(eig_vals)
var_exp=[(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
cum_var_exp=np.cumsum(var_exp)
print('Cumulative Explained Variance: \n',cum_var_exp)
```

```python
# Finding the number of components to be used to get 90% variance of the data:
len(cum_var_exp[cum_var_exp>91])
```

```python
df_sc.shape
```

```python
# Since the dataset has 31 columns after cleaning, 31-17 gives 14.
# Therefore, 14 PCA components are necessary to get the variance of 90% of the data.
```

```python
# C) Building a df with PCA components and check multi collinearity for the same.
from sklearn.decomposition import PCA
pca=PCA()
pca=PCA(n_components=14)
pca_df=pd.DataFrame(pca.fit_transform(df_sc))
pca_df.head(2)
```

```python
pca_df.corr().round(2)
```

```python
# From the correlation matrix above, it is evident that the multi collinearity has been totally removed by applying PCA.
```

```python
# D)
plt.figure(figsize=(10,7))
plt.bar(range(31),var_exp,alpha=0.5,align='center',label='Individual Explained Variance')
plt.step(range(31),cum_var_exp,where='mid',label='Cumulative Explained Variance')
plt.xlabel('No_of_Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
# E)
for i in pca_df.columns:
    sns.boxplot(pca_df[i])
    plt.show()
```

```python
# Performing outlier Treatment by capping/Winsorization Technique since the PCA components also has many outliers.

for i in pca_df.columns:
    q1=pca_df[i].quantile(0.25)
    q3=pca_df[i].quantile(0.75)
    iqr=q3-q1
    ub=q3+1.5*iqr
    lb=q1-1.5*iqr
    uc=pca_df[i].quantile(0.99)
    lc=pca_df[i].quantile(0.01)
    for ind1 in pca_df[i].index:
        if pca_df.loc[ind1,i]>ub:
            pca_df.loc[ind1,i]=uc
        if pca_df.loc[ind1,i]<lb:
            pca_df.loc[ind1,i]=lc
```

```python
# Checking outliers after capping:
for i in pca_df.columns:
    sns.boxplot(pca_df[i])
    plt.show()
```

```python
# The boxplot shows that the outliers are greatley reduced after capping.
```

```python
# A)
from sklearn.cluster import KMeans
wcsse=[]
for i in range(1,10):
    km=KMeans(n_clusters=i,random_state=0)
    km.fit(pca_df)
    wcsse.append(km.inertia_)
print(wcsse)
```

```python
df_clusters=pd.DataFrame({'cluster_range':range(1,10),'error':wcsse})
plt.plot(df_clusters['cluster_range'],df_clusters['error'],marker='o')
plt.show()
```

```python
# From the above, it is clear that n_clusters can either be 2 or 3. Hence, trying to build a model with n_cluster =2 and 
# building a KMeans clustering model and visualsing the spread of data for it.

from sklearn.cluster import KMeans
km=KMeans(n_clusters=2,random_state=0)
km.fit(pca_df)
labels_km=km.predict(pca_df)
```
